```python
import spacy
import random
import string
import numpy as np
import pandas as pd
from spacy import displacy
from spacy.training import offsets_to_biluo_tags
from spacy.training import Example
from spacy.tokens import DocBin, Doc
from spacy.util import minibatch, compounding
```

```python
df = pd.read_csv("synthetic_data.csv")

```

```python
df.head()
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
texts = df['Text'].tolist()
company_names = df['CompanyName'].tolist()
```

```python

ner_results = []

# Iterate over each text in the 'Text' column
for text in df['Text']:
    # Apply the NER organization model on the text
    doc = nlp(text)
    
    # Extract the organization entities from the document
    org_entities = [ent.text for ent in doc.ents if ent.label_ == 'ORG']
    
    # Append the organization entities to the results list
    ner_results.append(org_entities)
```
